```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('data.csv', header = None)
data.columns = ['x1', 'x2', 'label']

X = data[['x1', 'x2']].values
y = data['label'].values

# Plot data
plt.figure(figsize=(8, 6))
for label in set(y):
    plt.scatter(X[y == label][:, 0], X[y == label][:, 1], label=f'Class {label}')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Data from data.csv')
plt.show()
```

```python
import numpy as np

def heuristic_perceptron(X, y, learning_rate=0.1, max_epochs=20):
    n_samples, n_features = X.shape
    w = np.random.randn(n_features)
    b = np.random.randn()
    lines = []

    for epoch in range(max_epochs):
        for i in range(n_samples):
            x_i = X[i]
            y_pred = 1 if np.dot(w, x_i) + b >= 0 else 0
            error = y[i] - y_pred

            # Update weights and bias
            w += learning_rate * error * x_i
            b += learning_rate * error

        # Save line for plotting
        lines.append((w.copy(), b))

    return w, b, lines
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def gradient_descent_perceptron(X, y, learning_rate=0.1, epochs=100):
    n_samples, n_features = X.shape
    w = np.random.randn(n_features)
    b = np.random.randn()
    lines = []
    losses = []

    for epoch in range(epochs):
        epoch_loss = 0
        for i in range(n_samples):
            z = np.dot(w, X[i]) + b
            y_hat = sigmoid(z)
            error = y[i] - y_hat

            # Gradient updates
            w += learning_rate * error * X[i]
            b += learning_rate * error

            epoch_loss += - (y[i]*np.log(y_hat + 1e-9) + (1 - y[i]) * np.log(1 - y_hat + 1e-9))

        losses.append(epoch_loss / n_samples)
        lines.append((w.copy(), b))

    return w, b, lines, losses
```

```python
def plot_lines(X, y, lines):
    plt.figure(figsize=(8, 6))
    for label in set(y):
        plt.scatter(X[y == label][:, 0], X[y == label][:, 1], label=f'Class {label}')

    x_vals = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)

    for i, (w, b) in enumerate(lines):
        y_vals = -(w[0] * x_vals + b) / w[1]
        if i == 0:
            plt.plot(x_vals, y_vals, 'r-', label='Initial')
        elif i == len(lines) - 1:
            plt.plot(x_vals, y_vals, 'k-', label='Final')
        else:
            plt.plot(x_vals, y_vals, 'g--', alpha=0.4)

    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.title('Heuristic Perceptron Decision Boundaries')
    plt.show()

# Run it
final_w, final_b, lines = heuristic_perceptron(X, y, learning_rate=0.1)
plot_lines(X, y, lines)

# Run model
final_w_gd, final_b_gd, lines_gd, losses = gradient_descent_perceptron(X, y, learning_rate=0.1, epochs=100)

# Plot decision lines
plot_lines(X, y, lines_gd)

# Plot loss
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(losses) + 1), losses)
plt.xlabel('Epoch')
plt.ylabel('Log Loss')
plt.title('Log Loss over Epochs')
plt.grid()
plt.show()
```
